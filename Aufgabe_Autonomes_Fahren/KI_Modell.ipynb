# Szenario 3
**Bekannte Fakten:** F (Fußgänger erkannt) und A (Ampel ist rot)

```python
#Szanrio 3: Fußgänger und rote Ampel
bekannte_fakten = {AtomareFormel("F"), AtomareFormel("A")}  # F und A sind wahr: Fußgänger erkannt und Ampel ist rot
inferenz = wende_modus_ponens_an(wb, bekannte_fakten)
print("Abgeleitete Fakten:", inferenz)
```
